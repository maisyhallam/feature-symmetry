```python
import csv # for writing dataframes to csv
import random # for making a random choice
import os # for scanning directories
import itertools
import string # for generating strings
from collections import Counter

import kintypes as kt # bringing large lists of kin types into the namespace
import math # for calculating logs
import pandas as pd
import numpy as np
import re
from tqdm import tqdm
```

```python
pairs = kt.ics_pairs

relatives_gn = []
relatives_gn1 = []

for pair in pairs:
    relatives_gn1.append(pair[0])
    relatives_gn.append(pair[1])
    
relatives_gn
```

NB we want to get the meanings for Generation +1 and the meanings for Generation 0 separately because G0 should inherit features from their parents.

```python
def g1_meanings(relatives):
    relatives_features = {}
    
    for relative in relatives:
        
        features = []
        if 'M' in relative:
            features.append('mo')
        elif 'F' in relative:
            features.append('fa')
        else:
            features.append('na')

        if 'e' in relative:
            features.append('e')
        elif 'y' in relative:
            features.append('y')
        else:
            features.append('na')

        if 'Z' in relative:
            features.append('fem')
        elif 'B' in relative:
            features.append('male')
        else:
            if relative == 'mM' or relative == 'fM':
                features.append('fem')
            else:
                features.append('male')
                        
        relatives_features[relative] = features
    
    return relatives_features
```

```python
def g0_meanings(g1,pairs):
    features = {}
    
    for pair in pairs:
        bundle = []
        if pair[1] in features:
            pass
        else:
            bundle = g1_meanings(g1)[pair[0]]
            
            if 'Z' in pair[1][-1] or 'D' in pair[1][-1]:
                bundle.append('fem')
            elif 'B' in pair[1][-1] or 'S' in pair[1][-1]:
                bundle.append('male')

            if 'e' in pair[1][-2]:
                bundle.append('e')
            elif 'y' in pair[1][-2]:
                bundle.append('y')
            else:
                bundle.append('na')
                
            features[pair[1]] = bundle

    return features 
```

```python
g1 = g1_meanings(relatives_gn1)

g0 = g0_meanings(g1,pairs)

features = {**g1, **g0}

relatives = relatives_gn + relatives_gn1

print(relatives)

similarities = []

for relative1 in relatives:
    for relative2 in relatives:
        print(relative1)
        print(relative2)
        a = np.array(features[relative1])
        b = features[relative2] + ['na','na']
        b = np.array(b)
        
        if len(a) == len(b):
            similarity = np.sum(a == b)
            
        similarities.append(similarity)

print(similarities)
```
